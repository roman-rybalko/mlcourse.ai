```
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(1, 100, 1000)
```

```
def plot(y):
  plt.plot(y)
  sm.graphics.tsa.plot_acf(y)
  sm.graphics.tsa.plot_pacf(y)
  plt.show()
```

```
plot(np.sin(x))
```

```
plot(np.sin(x) + np.sin(x/2))
```
