```
import numpy
def func(x):
  return 3*pow(x,4)-8*pow(x,3)+6*pow(x,2)-12
x = numpy.arange(-3,3,0.0001)
min(func(x) for x in x)
```

```
import matplotlib.pyplot as plt
plt.plot(x, [func(x) for x in x])
plt.show()
```

```
def fibonacci(max):
  p = 0
  n = 1
  while n < max:
    yield n
    n, p = n + p, n
[x for x in fibonacci(100)]
```
